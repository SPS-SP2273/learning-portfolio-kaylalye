### Task 6

### Task 7

```python
cities_list = ['Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities_list:
    old_paths = glob.glob(f'*/*{city.lower()}.*')
    for files in old_paths:
        new_path = os.path.join(files.split(os.path.sep)[0], city, files.split(os.path.sep)[1])
        shutil.move(files, new_path)
        print(f'Moved {files.split(os.path.sep)[1]} to {city} folder')
```
