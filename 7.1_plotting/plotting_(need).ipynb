# 1 A simple plot

## 1.1 Letâ€™s look at some code.

### Things to note

```python
#data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10 ,12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label = 'Y values',
         color = 'red', linestyle = 'dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25) #changes the opacity of grid lines)
plt.legend(loc='upper left')
```

## 1.2 Adding another plot

```python
#data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10 ,12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label = 'Y1 values',
         color = 'red', linestyle = 'dashed')
plt.plot(x, y_2,
         label = 'Y2 values', c = 'blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25) #changes the opacity of grid lines)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
#data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10 ,12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, label = 'Y1 values',
         color = 'red', linestyle = 'dashed')
plt.plot(x, y_2,
         label = 'Y2 values', c = 'blue')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
         label = 'Y3 with errors', color = 'green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25) #changes the opacity of grid lines)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
import numpy as np

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10 ,12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, c='red', ls='dashed', label = 'Y1 Values')
plt.plot(np_x, np_y_2, c='blue', label='Y2 Values')
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25)
plt.legend(loc='upper left')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2', c='red', ls='dashed')
plt.plot(x, sin_x, label='sin(x)', c='blue')
plt.legend()
```

## We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2', c='red', ls='dashed')
plt.legend(loc='lower left')

plt.twinx()
#creates a new y-axis for the next plot

plt.plot(x, sin_x, label='sin(x)', c='blue')
plt.legend(loc='lower right')
```

### Things to note

```python
x = np.linspace(0, 6, 100)
#creates evenly spaced values over the range given
#np.linspace(start, stop, number of points)

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2', c='red', ls='dashed')
plt.legend(loc='lower left')

plt.twinx()
#creates a new y-axis for the next plot

plt.plot(x, sin_x, label='sin(x)', c='blue')
plt.legend(loc='lower right')
```

# 3 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2', c='red', ls='dashed')
plt.legend(loc='lower left')

plt.twinx()

plt.plot(x, sin_x, label='sin(x)', c='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24) #skips the rows in the beginning eg for headers
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
#alpha softens the colour of the plot
plt.ylim([-8, 8])
```

```python
plt.hlines(0, date[0], date[-1], ls='--', c='grey')
#needs a y-values and starting and ending x-values
```
