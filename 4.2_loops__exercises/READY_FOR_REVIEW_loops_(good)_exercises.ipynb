## Exercise 1

```python
%%timeit
list = []

for number in range (10):
    if number%2 == 0:
        continue
    else:
        list += [number**2]

print(list)
```

## Exercise 2

```python
%%timeit
[number**2 for number in range(10) if number%2 == 1]
```

## Exercise 3
Using list comprehension in Exercise 2 was faster, where the mean time, 1.61s, was shorter than that in Exercise 1, which was 474s.

## Exercise 4

```python
amount = 100
years = 0

while amount > 50:
    amount = amount*0.95
    years = years + 1

print(f"The sample will take {years} years to reduce to half its original amount.")
```
